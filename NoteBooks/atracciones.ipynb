```python
#Vamos a sustituir los valores nulos de la columnna comienzo de atracción. Y los valores que superen el tiempo de toma de los datos.
#Para ello crearemos una lista aleatoria con el total del tiempo - 1 de los datos
#utilizamos un bucle for con una selección aleatoria para sustituir los valores nulos. 
tiempo = [random.randint(1, 704) for i in range(96)]
for i in range(len(df_atracciones)):
    if pd.isnull(df_atracciones.loc[i,'comienzo_atraccion']):
         df_atracciones.loc[i, 'comienzo_atraccion'] = random.choice(tiempo)
    if df_atracciones.loc[i,'comienzo_atraccion']>705:
         df_atracciones.loc[i, 'comienzo_atraccion'] = random.choice(tiempo)
        
```

```python
#Estamos creando una lista con las atracciones.
atr=df_atracciones['atraccion'].unique()
lista_atracciones=list(atr)
lista_atracciones = [x for x in lista_atracciones if pd.notna(x)]
#lista_atracciones.pop()#una solución un poco chapucera. No entiendo porque no me funciona lo de arriba.
```

```python
lista_atracciones
```

```python
#Vamos a susutituir los valores nulos de la columna atracción mediante una selección aleatoria,
#para ello utilizaremos un bucle for. Utilizaremos la lsita creada anteriormente.
for i in range(len(df_atracciones)):
    if pd.isnull(df_atracciones.loc[i, 'atraccion']):
        df_atracciones.loc[i, 'atraccion'] = random.choice(lista_atracciones)
```

```python
#De esta manera quitamos el valor str "nan" que se nos quedó de la interación anterior, al no haber eliminado "nan" de la lista.
df_atracciones['atraccion'] = df_atracciones['atraccion'].replace("nan", random.choice(lista_atracciones))
```

```python
df_atracciones['comienzo_atraccion']=df_atracciones['comienzo_atraccion'].abs()
```
