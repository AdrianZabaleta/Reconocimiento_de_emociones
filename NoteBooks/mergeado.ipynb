```python
from functools import reduce

# Lista de DataFrames para fusionar. Hemos teniado que hacerlo solo con tres porque se bloqueaba el kernel
dfs = [df_procedencia, df_duracion, df_ticket]

# Usar reduce para fusionar todos los DataFrame de la lista
#Hay que tener en cuenta que la columna 't_id' está en todos los df que queremos mergear, por lo que esto puede ser un problema.
#Para solucionarlo  especifico suffixes=('', '_right') para evitar la colisión de nombres en las columnas, 
#asignando el sufijo '_right' solo a las columnas del DataFrame "right" en cada fusión. 

merge_by_id_visitante = reduce(
    lambda left, right: pd.merge(left, right, on='id_visitante', suffixes=('', '_right')), 
    dfs
)
# Eliminar columnas duplicadas que terminen con '_right'
merge_by_id_visitante = merge_by_id_visitante.loc[:, ~merge_by_id_visitante.columns.str.endswith('_right')]
```

merge_by_id_visitante = merge_by_id_visitante.loc[:, ~merge_by_id_visitante.columns.str.endswith('_right')]

-Componentes del Código

1. merge_by_id_visitante:

Este es el DataFrame que contiene el resultado de todas las fusiones realizadas anteriormente con la función reduce.

2. .loc[]:

Este es un accesor indexador usado en pandas para acceder a un grupo de filas y columnas por etiquetas o una matriz booleana. En este caso, .loc[:, ...] se utiliza para acceder a todas las filas (:) y a columnas específicas basadas en la condición que se define dentro de los corchetes.

3. merge_by_id_visitante.columns:

Esto devuelve un índice que contiene las etiquetas de las columnas del DataFrame merge_by_id_visitante.

4. .str.endswith('_right'):

Este es un método de cadena que se aplica a cada etiqueta de columna en el DataFrame. El método endswith('_right') devuelve True para las columnas cuyos nombres terminan con el sufijo _right, y False para los que no.
5. ~ (operador de negación):

Este operador invierte el resultado del filtro. Es decir, transforma True en False y viceversa. Por lo tanto, ~merge_by_id_visitante.columns.str.endswith('_right') genera una matriz booleana que es True para todas las columnas que no terminan en _right.

-Función del Código

El objetivo de esta línea es eliminar todas las columnas del DataFrame merge_by_id_visitante que terminan con _right. Esto es útil para deshacerse de las columnas duplicadas que se añadieron durante la fusión para evitar conflictos de nombres, dejando solo las columnas originales y las necesarias para el análisis o procesamiento posterior.


El uso de loc con la matriz booleana asegura que solo se seleccionen las columnas que no fueron marcadas con _right, eliminando efectivamente las columnas duplicadas o innecesarias del DataFrame final.


Este enfoque es muy práctico en procesos de fusión múltiple donde queremos mantener solo una versión de cada columna y evitar redundancias o conflictos en los datos.
