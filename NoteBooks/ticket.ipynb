```python
df=pd.read_csv('/Users/h.adrianzabaletahernandez/Desktop/ArchivosCSVProyecto-Pontia2024/ticket_sql.csv')
df
```

```python
#vamos a susutuir los valores erróneos mediante una lista y una selección aleatoria 
antelacion = [f"opcion_{i}" for i in range(0, 3)]
for i in range(len(df)):
    if df.loc[i, 'tipo_entrada'] == 'Pase Rápido' and df.loc[i, 'antelacion_de_compra'] > 3:
        df.loc[i, 'antelacion_de_compra'] = np.random.choice(antelacion)
```

```python
df.to_csv('ticket_sql_2.csv', index=False)
```
